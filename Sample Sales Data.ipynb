```python
data_filtered = data[data["YEAR_ID"] == 2003]
product_counts = data_filtered["PRODUCTCODE"].value_counts()
```

```python
top_five_products = product_counts.nlargest(5)
```

```python
plt.bar(top_five_products.index, top_five_products.values)
plt.xlabel("Product Name")
plt.ylabel("Times")
plt.title(f"Top 5 Products by Sold in 2003")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
